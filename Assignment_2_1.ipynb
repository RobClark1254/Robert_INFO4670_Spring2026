#2. Data Cleaning and Preprocessing

2.1 Remove the first column or “Column 1” from the dataset. Treat the ‘year’ variable as nominal.

```python
df = df.drop(columns=['Column 1'])

df['year'] = df['year'].astype('category')

print("Dataset after removing 'Column 1' and converting 'year' to category:")
display(df.head())
print("Data types after changes:")
display(df.dtypes)
```

2.2 Check for duplicate values and remove them.

```python
num_duplicate_rows = df.duplicated().sum()

if num_duplicate_rows > 0:
    print(f"Found {num_duplicate_rows} duplicate rows. Removing them...")
    df.drop_duplicates(inplace=True)
    print("Duplicate rows removed successfully.")
    print(f"New number of rows: {len(df)}")
else:
    print("No duplicate rows found.")

num_duplicate_rows_after = df.duplicated().sum()
if num_duplicate_rows_after == 0:
    print("Confirmed: No duplicate rows remaining.")
else:
    print(f"Warning: {num_duplicate_rows_after} duplicate rows still exist after removal attempt.")
```

2.3 Check for missing values. If a data record (row) only has a few missing values, replace the missing values with the median of the column feature in that specific “Region” variable. If most column values in a data record are missing, remove the data record.

```python
print("\n    Missing Values Before Handling    ")
missing_values_count = df.isnull().sum()
missing_values_count = missing_values_count[missing_values_count > 0]
if not missing_values_count.empty:
    print(missing_values_count)
else:
    print("No missing values found.")

df['non_missing_count'] = df.count(axis=1)
num_columns_for_consideration = len(df.columns) - 1

#Threshhold for what counts as missing too much info
missing_threshold = num_columns_for_consideration * 0.5

rows_to_drop = df[df['non_missing_count'] < (num_columns_for_consideration - missing_threshold)].index

if not rows_to_drop.empty:
    print(f"\nDropping {len(rows_to_drop)} rows where most values are missing...")
    df.drop(rows_to_drop, inplace=True)
    print(f"Remaining rows: {len(df)}")
else:
    print("\nNo rows found with most values missing (based on threshold).")

df.drop(columns=['non_missing_count'], inplace=True)

print("\n--- Imputing remaining missing values by Region ---")

numerical_cols = df.select_dtypes(include=['number']).columns

for col in numerical_cols:
    if df[col].isnull().any():
        df[col] = df.groupby('region')[col].transform(lambda x: x.fillna(x.median()))
        print(f"Imputed missing values in '{col}' with median by region.")

categorical_cols = df.select_dtypes(include=['object', 'category']).columns
for col in categorical_cols:
    if df[col].isnull().any():
        df[col] = df.groupby('region')[col].transform(lambda x: x.fillna(x.mode()[0] if not x.mode().empty else x.dropna().sample(1).iloc[0] if not x.dropna().empty else 'Unknown'))
        print(f"Imputed missing values in categorical column '{col}' with mode/fallback by region.")

print("\n--- Missing Values After Handling ---")
missing_after_handling = df.isnull().sum()
missing_after_handling = missing_after_handling[missing_after_handling > 0]
if not missing_after_handling.empty:
    print("Warning: Some missing values still exist:")
    print(missing_after_handling)
else:
    print("All missing values have been handled successfully.")

print("\nUpdated DataFrame head after handling missing values:")
display(df.head())
```
